```python
import cv2
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample.jpg")
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
display(gray)
```

```python
# plt.figure(figsize=(12,10))
plt.imshow(gray,cmap='gray')
# plt.imshow(gray)
plt.title("Original Grayscale Image")
plt.axis("off")
```

```python
improved_brightness_image = cv2.add(gray,100)
plt.imshow(improved_brightness_image,cmap='gray')
plt.axis("off")
plt.title("Image With Brightness Improvement")
```

```python
reduced_brightness_image = cv2.subtract(gray,100)
plt.imshow(reduced_brightness_image,cmap='gray')
plt.axis("off")
plt.title("Image With Brightness Reduction")
```

```python
_, thresholded_image = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded_image,cmap='gray')
plt.title("Image After Thresholding")
plt.axis("off")
```

```python
# Using cv2.bitwise_not

negative_image = cv2.bitwise_not(gray)
plt.imshow(negative_image,cmap='gray')
plt.title("Negative Image")
plt.axis("off")
```

```python
# Using Formula
negative_image = 255 - gray 
plt.imshow(negative_image,cmap='gray')
plt.title("Negative Image")
plt.axis("off")
```

```python
gray_float = gray.astype(np.float32)

c = 255 / np.log(1 + np.max(gray_float)) 
log_transformed_image = c * np.log(1 + gray_float)

log_transformed_image = np.array(log_transformed_image, dtype=np.uint8)

plt.imshow(log_transformed_image, cmap='gray')
plt.title("Log Transformed Image")
plt.axis("off")
```

```python
gamma = 0.5

gray_float = gray.astype(np.float32) / 255.0

power_law_transformed_image = cv2.pow(gray_float, gamma)

power_law_transformed_image = np.uint8(power_law_transformed_image * 255)

plt.imshow(power_law_transformed_image, cmap='gray')
plt.title("Image After Power Law Transformation")
plt.axis("off")
```
